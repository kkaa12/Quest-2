```python
import numpy as np

def standardize_activity(activity):
    if activity is np.nan:
        return 'Unknown'
    activity = activity.lower().strip()
    if 'surf' in activity:
        return 'Surfing'
    elif 'swim' in activity or 'bathing' in activity or 'wading' in activity or 'treading' in activity:
        return 'Swimming'
    elif 'diving' in activity or 'snorkel' in activity:
        return 'Diving'
    elif 'fishing' in activity or 'fish' in activity:
        return 'Fishing'
    elif 'kayak' in activity or 'canoe' in activity:
        return 'Kayaking/Canoeing'
    elif 'boarding' in activity or 'paddle' in activity:
        return 'Board Sports'
    elif 'sitting' in activity or 'standing' in activity:
        return 'Stationary Activities'
    elif 'walking' in activity or 'hiking' in activity:
        return 'Walking/Hiking'
    else:
        return 'Other'


df['activity'] = df['activity'].apply(standardize_activity)
```

```python
import re #library that allows you to search, match, and find all occurrences of specified patterns within strings

def simple_clean_age(value):
    if pd.isnull(value):
        return np.nan

    value = str(value)

    if 'teen' in value.lower():
        return 15
    elif 'adult' in value.lower():
        return 30
    elif 'elderly' in value.lower():
        return 70
    elif 'young' in value.lower():
        return 20

    match = re.search(r'\d+', value)
    return float(match.group()) if match else np.nan

df['age'] = df['age'].apply(simple_clean_age)
```

```python
df['age'] = pd.to_numeric(df['age'])
```

```python
# Filter our data, as we focus on the years from 2000 to 2024

start_year = 2000
end_year = 2024

filtered_df = df[(df['year'] >= start_year) & (df['year'] <= end_year)]
filtered_df['year'].unique()
```

```python
filtered_df.groupby('year').size()
```

```python
# Calculate year with most shark attacks per year

attacks_per_year = filtered_df.groupby('year').size()

sorted_attacks = attacks_per_year.sort_values(ascending=False)
year_most_attacks = sorted_attacks.index[0]
most_attacks = sorted_attacks.iloc[0]

print(f"The year with the most shark attacks is {year_most_attacks} with {most_attacks} attacks.")
year_most_attacks
```

```python
# Percentage of shark attacks that end in death

num_fatal = filtered_df['injury_fatal'].value_counts().get('Fatal', 0)

total_attacks = len(filtered_df)

if total_attacks > 0:  
    fatal_percentage = (num_fatal / total_attacks) * 100
else:
    fatal_percentage = 0

print(f"The percentage of shark attacks that ended in death is {fatal_percentage:.2f}%.")
```
